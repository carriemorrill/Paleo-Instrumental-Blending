This Jupyter notebook uses some simplified "quickstart" code from the LMR to calculate a reconstruction of annual-mean surface air temperature for one year (1800 CE) using:  
  
a) the PAGES2k version 2 proxy database expanded with Anderson et al. 2019 database  
b) a simple Proxy System Model (PSM) relating proxy measurements and climate observations (GISTEMP temperature, GPCC precipitation) via linear regression over a 1850-2015 CE period of overlap  
c) simulated temperature and precipitation fields from a last millennium simulation completed with the Community Climate System model version 4 (CCSM4).

## 4. Load pre-calculated PSM values
The "Proxy System Model" (PSM) is the relationship between proxy values (e.g., tree rings, coral d18O) and observed climate variables (temperature, precipitation) as defined by a linear regression over a 1850-2015 period of overlap (i.e., the calibration period). This linear regression (climate variable vs. proxy value) is calculated beforehand by other code, and has also been used to "forward model" proxy values for each year of the model prior. This forward modeling is achieved by applying the linear regression relationship to each year of climate model output to "simulate" the values a proxy would have for that year. Note that since we randomly chose 100 years of climate model output above, we are loading PSM results for just those same 100 years below.

```python
Ye_assim, Ye_assim_coords = LMR_utils.load_precalculated_ye_vals_psm_per_proxy(cfg, prox_manager,'assim',X.prior_sample_indices)
```

## 5. Data assimilation for one year (creates the "posterior")
This is how all the various parts (proxy measurements, model prior, proxy system model) fit together to do the data assimilation (i.e., generate the "posterior"). The data assimilation is based on the following general equation:  

$Xa = Xb + K[y-ye]$  

where $Xa$ is the posterior, $Xb$ is the prior, $K$ is the Kalman gain matrix, $y$ is the proxy measurements and $ye$ is the proxy system model estimates. The Kalman gain matrix is calculated from information about the prior, the proxy system model, and the proxy measurements. The Kalman gain matrix spreads $[y-ye]$, or the "innovation", in space based on spatial covariances in the model (the degree to which climate in one place is related to climate elsewhere) and also scales the innovation based on the amount of error in the proxy measurements. For example, as error in the proxy measurements increases, the impact of proxy measurements on altering the prior into the posterior decreases. 

```python
target_year=1800
print('performing a test reconstruction for year:' + str(target_year))
vY,vR,vP,vYe,vT,vYe_coords = LMRlite.get_valid_proxies(cfg,prox_manager,target_year,Ye_assim,Ye_assim_coords)
# vY = Yobs; vR = obs.err; vP = proxy.idx; vYe = precalc PSM; vT = Y.type; vYe_coords = precalc PSM lat & lon
xam,Xap,_ = LMRlite.Kalman_optimal(vY,vR,vYe,Xb_one,verbose=False)
```

# Other notes
1. To do a robust data assimilation, we would repeat steps #2-5 many more (~50) times and pool the results. This "Monte Carlo" sampling over subsets of model priors and proxy datasets makes the results more accurate.
2. Variables other than temperature can be reconstructed by adding these variables (e.g., SPEI) as priors.
3. Benefits of data assimilation over more traditional reconstruction methods include (a) the ability to simultaneously estimate hydroclimate, temperature, and other aspects of the atmosphere-ocean state and (b) solutions that are more physically consistent since dynamically-constrained climate model fields are used.
4. Possible future improvements include (a) bias correcting the prior (e.g., climate models often "run" too hot/cold or wet/dry in some areas and these biases can be removed before data assimilation), (b) refined proxy system models that explain more proxy variability, (c) expanded proxy networks with newly-generated records.
